```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
print("komal chilke\nBE21F06F014")

dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/advertising.csv")
print ("\nDataset Head: \n")
print (dataset.head())
x = dataset [['TV', 'Radio', 'Newspaper']]
y = dataset ['Sales']
print ("\nIndependent Variables (X): \n")
print (x)

print ("\nDependent Variable (y): \n")
print (y)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.3, random_state = 100)
print ("\nTraining Data (x):\n")
print (x_train)
print ("\nTraining Data (y):\n")
print (y_train)
print ("\nTest Data (x):\n")
print (x_test)
print ("\nTest Data (y) :\n")
print (y_test)

multiple_LR = LinearRegression()
multiple_LR.fit(x_train, y_train)

print ("\nIntercept: ", multiple_LR. intercept_)
print ("\nCoeï¬ƒcients:")
print (list (zip(x, multiple_LR.coef_)))

y_pred_mlr= multiple_LR.predict(x_test)

print ("\nPrediction for test set: \n\n {}".format (y_pred_mlr))

mlr_diff = pd.DataFrame ({'Actual value': y_test, 'Predicted value': y_pred_mlr})
print ("\nActual Vs Predicted: \n",mlr_diff.head())

from sklearn import metrics
meanAbErr = metrics.mean_absolute_error (y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error (y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error (y_test, y_pred_mlr))

print("\nModel Evaluation: ")
print('R squared: {:.2f}'.format (multiple_LR. score (x, y) *100))
print('Mean Absolute Error (MAE):', meanAbErr)
print('Mean Square Error (MSE): ', meanSqErr)
print('Root Mean Square Error (RMSE): ', rootMeanSqErr)
```
