```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
dataset = pd.read_csv("/content/p7Dataset.csv")
dataset.head()
```

```python
X = dataset.iloc[:, 2:4]
y = dataset.iloc[:, -1]
```

```python
print ("Input Features: \n")
X
```

```python
print("Output/Target Feature:\n")
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size= 0.20, random_state = 0)
```

```python
print("Training Data (Input):\n")
X_train
```

```python
print ("Training Data (Output): \n")
y_train
```

```python
print ("Test Data (Input):\n")
X_test
```

```python
print("Test Data (Output): \n")
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform (X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB ()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix (y_test, y_pred)
print("Confusion Matrix: \n", cm)
```

```python
ac= accuracy_score(y_test,y_pred)
print("Accuracy of model =",ac*100)
```
