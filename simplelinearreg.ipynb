```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams ['figure.figsize'] = (7.0, 4.0)
print("komal chilke\nBE21F06F014")

data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML_1_data.csv")
print ("\nDataset: ")
print (data)
X = data.iloc[:, 0]
Y = data.iloc[:, 1]

print ("\nIndependent Variables\n")
print (X)
print ("\nDependent Variables")
print (Y)
print ("\n\n")
print ("\nScatter lot of X and Y")
plt.scatter (X, Y)
plt.show()
print ("\n\n")
X_mean = np.mean(X)
Y_mean = np.mean(Y)
num = 0
den = 0
for i in range (len (X)):
  num += (X[i] - X_mean)*(Y[i] - Y_mean)
  den += (X[i] - X_mean)**2
m = num / den
c = Y_mean - m*X_mean
print ("Values of slope and intersect are: ")
print ("m={}, c={}".format (m, c))
Y_pred = m*X + c
print ("\n\nBest fit line")
plt.scatter (X, Y)
plt.scatter (X, Y_pred, color='red')
plt.plot([min(X), max(X)], [min(Y_pred), max (Y_pred)], color='blue')
plt.show()

```
